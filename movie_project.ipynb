```python

#13 Identify the 3 most frequent actors in the dataset
all_actors = trex['Actors'].str.cat(sep=', ')
actor_list = all_actors.split(', ')
actor_counts = pd.Series(actor_list).value_counts()
top_actors = actor_counts.head(3)
print(top_actors)
```

```python

# 20 Calculate average ratings per year
average_ratings = trex.groupby('Year')['Rating'].mean()

# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(average_ratings.index, average_ratings.values, marker='o')
plt.title('Average Movie Ratings Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.grid(True)
plt.show()
```
